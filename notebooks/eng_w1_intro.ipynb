## Programming Languages: History and Popular Technologies

Definitions:

- Programming language: a set of commands and rules used to create a computer program
- Computer program: a sequence of instructions prepared in a programming language, translated into machine language
- Simplification: a computer program is an algorithm

Types of programming languages:
- Compiler: running a program after a change in the code requires translating the code into instructions understandable to the device (e.g. C++, Java)
- Scripting: the code uses ready-made functions that are already compiled (e.g. Python, PHP)

Generations of programming languages:
- 5GL: languages ​​that solve problems with no or minimal involvement of the programmer, e.g. Prolog (a logical programming language)
- 4GL: code that calls code written in 3GL, e.g. SQL, LabVIEW; graphical programming environments
- 3GL: more device agnostic, examples include C, C++, Java; in this group also more advanced languages ​​than 3GL, often using 3GL but not meeting the criteria of 4GL, e.g. Python
- 2GL: code in which one instruction corresponds to one instruction of machine code
- 1GL: machine code, specific to a given device

Usage of programming languages – sample division:
- Devices
- Operating systems
- Applications
- Websites
- Analytical and data processing tools

Usage of programming languages – other division:
- Frontend
- Backend
- Fullstack
- Mobile
- Embedded
- Data science
- Other

Types of programming languages:
- Structural: programming paradigm based on dividing the program's source code into procedures and hierarchically arranged blocks using control structures in the form of selection and loop instructions
- Object-oriented: programming paradigm in which programs are defined using objects – elements combining state and behavior
- Declarative: based on the assumption that the programmer defines the goal, not the way to achieve it
- Functional: emphasis is on evaluating functions, not executing commands
